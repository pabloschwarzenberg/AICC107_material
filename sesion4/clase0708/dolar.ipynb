```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("dolar.csv",sep=";")
```

```python
valores=df["valor"].to_numpy()
```

```python
valores
```

```python
x=[]
y=[]
historia=5
for i in range(len(valores)):
    if(i+historia<len(valores)):
        xi=[]
        for j in range(historia):
            xi.append(valores[i+j])
        x.append(xi)
        y.append(valores[i+historia])
```

```python
x
```

```python
y
```

```python
from keras.layers import Input, Dense
from keras.models import Sequential
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
import numpy as np
import datetime
import tensorflow as tf
```

```python
seed=12122008
x_train, x_val, y_train, y_val = train_test_split(
            x, y, test_size=0.2, random_state=seed)
```

```python
model = Sequential()
model.add(Input(name="historia",shape=(historia,)))
model.add(Dense(64))
model.add(Dense(1))
model.summary()
tag="mlp_serie"+str(i)+"-"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
log_dir = "logs/fit/" + tag
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
optimizer=Adam(learning_rate=1e-3)
model.compile(optimizer=optimizer, loss='mse', metrics=["mse","mae"])
```

```python
model.fit(x_train, y_train, validation_data=(x_val,y_val), batch_size=64, epochs=32,verbose=1,callbacks=[tensorboard_callback])
model.save('modelos/model_{0}.keras'.format(tag))
yv_pred=model.predict(x_val)
```

```python
fig = plt.figure()
fig.set_size_inches(32,10)
ax1 = fig.add_subplot(111)
eje=np.arange(0, len(x_val), 1.0)
plt.xticks(eje)

ax1.plot(eje, y_val, c='b', label='real')
ax1.plot(eje, yv_pred, c='r', label='pred')
plt.legend(loc='lower right')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(valores)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
x=[i for i in range(len(valores))]
fig, axes = plt.subplots(1,3)
fig.set_size_inches(15,5)
axes[0].plot(x,valores);
plot_acf(valores,lags=6,ax=axes[1]);
plot_pacf(valores,lags=6,method="ywm",ax=axes[2]);
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
modelo=ARIMA(valores,order=(1,0,0))
predictor=modelo.fit()
print("MSE",predictor.mse)
predictor.summary()
```
